```python
batting = pd.read_csv(r'C:\Users\ASUS\Desktop\Kaggle\Data Analysis\Cricket_Analysis\WebData\BattingSummary.csv')
```

```python
batting.head(10)
```

```python
dismissal_type = []
for wicket in batting['Out']:
    if wicket.startswith('c'):
        dismissal_type.append("Caught")
    elif wicket.startswith('b'):
        dismissal_type.append("Bowled")
    elif wicket.startswith('s'):
        dismissal_type.append("Stumped")
    elif wicket.startswith('lbw'):
        dismissal_type.append("LBW")
    elif wicket.startswith('run'):
        dismissal_type.append("Run Out")
    else:
        dismissal_type.append("Not Out")
```

```python
batting['Dismissal_Type'] = dismissal_type
```

```python
batting.drop(['Out'],axis=1,inplace=True)
```

```python
batting.info()
```

```python
#changing datatype 
batting['S.R'] = batting['S.R'].replace("-","0")

numeric_cols = ['Pos','Inning','Runs','Balls','Fours','Sixes','S.R']

for col in numeric_cols:
    batting[col] = pd.to_numeric(batting[col])
```

```python
batting.info()
```

```python
batting['Name'] = batting['Name'].apply(lambda x: x.replace('â€',''))
batting['Name'] = batting['Name'].apply(lambda x: x.replace('\xa0',''))
```

```python
batting.to_csv(r'C:\Users\ASUS\Desktop\Kaggle\Data Analysis\Cricket_Analysis\Processed_Data\batting.csv',index=False)
```
